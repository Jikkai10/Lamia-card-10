# Outliers

Vamos criar dados que envolvem a quantidade de dinheiro que tem a populaçao

Os dados serão dados por uma curva normal entorno de 27000

Mas sera colocado um bilionario


```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, [1000000000])

import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

Podemos ver que todo o resto da população ficou em um unico ponto por conta do bilionario

e se tirarmos a média, vemos o quão longe está dos 27000 da curva normal

```python
incomes.mean()
```


Nesse caso para termos uma ideia melhor é necessário filtrar os outliers

Vamos utilizar 2 desvios padroes para identificar outliers

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered

filtered = reject_outliers(incomes)

plt.hist(filtered, 50)
plt.show()
```

Agora os dados estão muito melhor dispostos

E a média também a mais realistica, em torno de 27000

```python
np.mean(filtered)
```
