Percebemos no material da aula que a maior parte da informação foi mantida em 1 dimensão

Por isso vamos testar fazer a redução parra apenas 1 dimensão

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pylab as pl

from itertools import cycle

iris = load_iris()

numSamples, numFeatures = iris.data.shape
print(numSamples)
print(numFeatures)
print(list(iris.target_names))

```

Treinando a PCA para 1 dimensão e transformando os dados

```python
X = iris.data
pca = PCA(n_components=1, whiten=True).fit(X)
X_pca = pca.transform(X)
```

Percebe-se que se manteve grande parte da informação

```python
print(pca.explained_variance_ratio_)
```

Com o gráfico podemos perceber que segue linearmente e basicamente muda conforme quão maior o valor

```python
%matplotlib inline
from pylab import *

colors = cycle('rgb')
target_ids = range(len(iris.target_names))
pl.figure()
for i, c, label in zip(target_ids, colors, iris.target_names):
    pl.scatter(X_pca[iris.target == i], X_pca[iris.target == i],
        c=c, label=label)
pl.legend()
pl.show()
```

Isso indica que provavelmente o tipo está fortemente relacionado ao tamanho total da iris
