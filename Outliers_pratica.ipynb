Agora vamos adicionar mais alguns bilionarios e trabalhar com eles

```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)

outliers = np.random.randint(100000000,1000000000, 100) 
incomes = np.append(incomes, outliers)




import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

```python
incomes.mean()
```

Com 2 desvios padrões percebemos que foram filtrados os mais ricos, mas ainda há a presença de outliers que dificultam a visualizaçào

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered

filtered = reject_outliers(incomes)

print(np.mean(filtered))
plt.hist(filtered, 50)
plt.show()
```

Com 1 desvio padrão o resultado já melhora, retirando os outliers

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 1 * s < e < u + 1 * s)]
    return filtered

filtered = reject_outliers(incomes)

print(np.mean(filtered))
plt.hist(filtered, 50)
plt.show()
```
